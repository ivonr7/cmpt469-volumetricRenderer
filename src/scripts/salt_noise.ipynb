```python
from src.benchmark_render import imentropy
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
file = 'llvm_output/render_samples_8_depth_16.png'
folder = 'cuda_output'
im = plt.imread(file)
plt.imshow(im)
```

```python
import os
import re
files = []
order = []
pat = re.compile(r"render_samples_(\d+)_depth_16")
for file in os.listdir(folder):
    if pat.match(file):
        files.append(os.path.join(folder,file))
        order.append(pat.groups()[0])
print(files[0])
ents = [imentropy(plt.imread(file)) for file in tqdm(files)]
```

```python
plt.plot(range(len(ents)),ents)
plt.xticks(range(len(ents)),labels=[f"{2**(i+1)}" for i in range(len(ents))],rotation = 45)
plt.title("Entropy vs SPP")
```
