```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
g = 0.4
g2 = 0.3
f = 0.7
x = np.linspace(-np.pi,np.pi,100)
hg = (1/4*np.pi) * (1-g**2) / (1+g - 2*g* np.cos(x))
schlick = (1 - g**2) / (4*np.pi * (1-g*np.cos(x)) ** 2)
dhg = f* ((1 - g**2) / (4*np.pi * (1-g*np.cos(x)) ** 2)) + (1-f) * ((1 - g2**2) / (4*np.pi * (1-g2* np.cos(x)) ** 2))
```

```python
plt.plot(x,hg,label = 'HG function')
plt.plot(x,dhg,label = 'DHG function g2=0.3')
plt.plot(x,schlick,label = 'Schlick function')
plt.legend()
plt.ylim((0,5))
plt.title(f"Comparison Between 3 Phase Functions (g={g})")
plt.xlabel("theta")
plt.ylabel("Response")
```
